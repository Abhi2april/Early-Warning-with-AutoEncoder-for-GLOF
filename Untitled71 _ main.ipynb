```python
# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the data and transform
scaled_values = scaler.fit_transform(df)

# Convert back to DataFrame for convenience
df_scaled = pd.DataFrame(scaled_values, index=df.index, columns=df.columns)

print("Data after normalization:")
print(df_scaled.head())
```

```python
df=df_scaled
```

```python
TIME_STEPS = 30  # We'll use 30 days as our sequence length
SPLIT_RATIO = 0.8  # 80% for training, 20% for testing
```

```python
def create_sequences(values, time_steps=TIME_STEPS):
    output = []
    for i in range(len(values) - time_steps + 1):
        output.append(values[i : (i + time_steps)])
    return np.stack(output)
```

```python
import numpy as np

def detect_anomalies(model, data, threshold):
    reconstructions = model.predict(data)
    reconstructions = np.squeeze(reconstructions)
    mse = np.mean(np.square(data - reconstructions), axis=1)
    return mse > threshold
```

```python
def plot_results(feature, original_data, anomalies, anomalous_data):
    plt.figure(figsize=(20, 10))

    plt.plot(original_data.index, original_data.values, label='Original', color='blue', alpha=0.5)

    plt.scatter(anomalous_data.index, anomalous_data.values, color='red', label='Anomaly', s=100, zorder=5)

    plt.title(f'Anomalies in {feature}', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Normalized Value', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```

```python
df_normalized = (df - df.mean()) / df.std()
```

```python
split_point = int(len(df) * SPLIT_RATIO)
train_data = df_normalized.iloc[:split_point]
test_data = df_normalized.iloc[split_point:]
```

```python
def print_anomaly_stats(feature, anomalies, original_data):
    print(f"\nStatistics for anomalies in {feature}:")
    print(f"Number of anomalies: {len(anomalies)}")
    if len(anomalies) > 0:  # Check if anomalies is not empty
        print(f"Percentage of anomalies: {len(anomalies) / len(original_data) * 100:.2f}%")
        print(f"Mean of anomalies: {anomalies.mean():.4f}")
        print(f"Std dev of anomalies: {anomalies.std():.4f}")
        print(f"Min anomaly value: {anomalies.min():.4f}")
        print(f"Max anomaly value: {anomalies.max():.4f}")
    else:
        print("No anomalies detected.")
    print(f"Original data mean: {original_data.mean():.4f}")
    print(f"Original data std dev: {original_data.std():.4f}")
```

```python
results = {}
```

```python
for feature in df.columns:
    print(f"Processing feature: {feature}")

    x_train = create_sequences(train_data[feature].values)

    model = keras.Sequential([
        layers.Input(shape=(x_train.shape[1], 1)),
        layers.Conv1D(filters=32, kernel_size=7, padding="same", activation="relu"),
        layers.Dropout(rate=0.2),
        layers.Conv1D(filters=16, kernel_size=7, padding="same", activation="relu"),
        layers.Conv1DTranspose(filters=16, kernel_size=7, padding="same", activation="relu"),
        layers.Dropout(rate=0.2),
        layers.Conv1DTranspose(filters=32, kernel_size=7, padding="same", activation="relu"),
        layers.Conv1DTranspose(filters=1, kernel_size=7, padding="same"),
    ])
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss="mse")

    history = model.fit(
        x_train, x_train,
        epochs=50,
        batch_size=128,
        validation_split=0.1,
        callbacks=[keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, mode="min")],
        verbose=0
    )

    reconstructions = model.predict(x_train)

    reconstructions = np.squeeze(reconstructions)

    train_loss = np.mean(np.square(x_train - reconstructions), axis=1)
    threshold = np.max(train_loss)

    x_test = create_sequences(test_data[feature].values)
    test_anomalies = detect_anomalies(model, x_test, threshold)

    anomalous_data_indices = []
    for data_idx in range(TIME_STEPS - 1, len(test_data) - TIME_STEPS + 1):
        if np.all(test_anomalies[data_idx - TIME_STEPS + 1 : data_idx]):
            anomalous_data_indices.append(data_idx)

    results[feature] = {
        'model': model,
        'threshold': threshold,
        'anomalies': test_data.iloc[anomalous_data_indices]
    }

    plot_results(feature, test_data[feature], test_anomalies, results[feature]['anomalies'])
    print_anomaly_stats(feature, results[feature]['anomalies'], test_data[feature])
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

def create_sequences(data, TIME_STEPS=100):
    """Create sequences for input into the model."""
    sequences = []
    for i in range(len(data) - TIME_STEPS + 1):
        sequences.append(data[i:i + TIME_STEPS])
    return np.array(sequences)

def custom_complex_model(sequence_length, num_features=1):
    inputs = keras.Input(shape=(sequence_length, num_features))

    x = layers.Conv1D(filters=64, kernel_size=7, padding="same")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.2)(x)
    encoder_skip1 = x

    x = layers.Conv1D(filters=32, kernel_size=5, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.2)(x)


    x = layers.LSTM(64, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)
    lstm_skip1 = x

    x = layers.LSTM(32, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)

    x = layers.LSTM(16, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)


    x = layers.LSTM(32, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)

    x = layers.LSTM(64, return_sequences=True)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, lstm_skip1])
    x = layers.Dropout(0.2)(x)


    x = layers.Conv1DTranspose(filters=32, kernel_size=5, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv1DTranspose(filters=64, kernel_size=7, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Add()([x, encoder_skip1])
    x = layers.Dropout(0.2)(x)

    outputs = layers.Conv1D(filters=1, kernel_size=1, padding="same")(x)

    model = keras.Model(inputs=inputs, outputs=outputs)

    def custom_loss(y_true, y_pred):
        mse = tf.keras.losses.mean_squared_error(y_true, y_pred)
        mae = tf.keras.losses.mean_absolute_error(y_true, y_pred)
        return mse + 0.5 * mae

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss=custom_loss,
        metrics=['mae']
    )

    return model

def detect_anomalies(model, data, threshold):
    reconstructions = model.predict(data)
    mse = np.mean(np.square(data - reconstructions), axis=(1, 2))
    return mse > threshold

def process_feature(train_data, test_data, feature, TIME_STEPS=100):
    scaler = StandardScaler()
    train_scaled = scaler.fit_transform(train_data[[feature]])
    test_scaled = scaler.transform(test_data[[feature]])

    x_train = create_sequences(train_scaled, TIME_STEPS)

    model = create_complex_model(TIME_STEPS)

    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor="val_loss",
            patience=5,
            mode="min",
            restore_best_weights=True
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=3,
            min_lr=0.00001
        )
    ]

    history = model.fit(
        x_train, x_train,
        epochs=50,
        batch_size=64,
        validation_split=0.2,
        callbacks=callbacks,
        verbose=1
    )

    reconstructions = model.predict(x_train)
    train_loss = np.mean(np.square(x_train - reconstructions), axis=(1, 2))
    threshold = np.mean(train_loss) + 3 * np.std(train_loss)

    x_test = create_sequences(test_scaled, TIME_STEPS)
    anomalies = detect_anomalies(model, x_test, threshold)

    anomaly_indices = []
    for i in range(len(anomalies)):
        if anomalies[i]:
            anomaly_indices.append(i + TIME_STEPS - 1)

    return {
        'model': model,
        'threshold': threshold,
        'anomaly_indices': anomaly_indices,
        'history': history.history
    }
```
